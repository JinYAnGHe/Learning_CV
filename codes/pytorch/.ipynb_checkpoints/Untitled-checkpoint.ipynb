```python
import warnings
warnings.filterwarnings('ignore')

%matplotlib inline
from pycocotools.coco import COCO
import numpy as np
import skimage.io as io
import matplotlib.pyplot as plt
import pylab
pylab.rcParams['figure.figsize'] = (8.0, 10.0)
```

```python
root_dir="/home/yang/Downloads/Competition/datasets/xray/traindata"
annFile='{}/train_no_poly.json'.format(root_dir)
```

```python
coco=COCO(annFile)
```

```python
print('cats:', len(coco.dataset['categories']))
print('images:', len(coco.dataset['images']))
print('annotations', len(coco.dataset['annotations']))
```

```python
# display COCO categories and supercategories
cats = coco.loadCats(coco.getCatIds())
# print(cats)
names = [cat['name'] for cat in cats]
print('Xray categories: \n{}\n'.format(' '.join(names)))
```

```python
counts = []
for name in names:
    catIds = coco.getCatIds(catNms=name)
    imgIds = coco.getImgIds(catIds=catIds)
    counts.append(len(imgIds))
print(names)
print(counts)
```

```python
plt.bar(range(len(counts)), counts)
plt.xlabel('categories')
plt.ylabel('conuts')
```

```python
img = coco.loadImgs(imgIds[np.random.randint(0, len(imgIds))])[0]

print('file_name: ', img['file_name'])

# load and display image
image = io.imread('%s/%s/%s'%(root_dir, 'restricted', img['file_name']))
plt.axis('off')
plt.imshow(image)
plt.show()
```

```python
# load and display instance annotations
import cv2
annIds = coco.getAnnIds(imgIds=img['id'], catIds=catIds, iscrowd=None)
anns = coco.loadAnns(annIds)
for ann in anns:
    x, y, w, h = ann['bbox']
    poly = ann['minAreaRect']
    p = np.asarray(poly)
    cv2.rectangle(image, (x, y), (x + w,y + h), (0, 0, 255), thickness=3)
plt.imshow(image)
plt.show()
```
